## Aufgabe 2
#### Welchen Wert muss $n \ge 0$ haben, damit $\sum \limits _{k=1} ^{n} \dfrac{2(nk+k)}{n} = 676 $ gilt?

```python
n= symbols("n")
k= symbols("k")
Sum(((2*(n*k+k))/n),(k,1,n)).doit() 
#Ergebnis der Summe ausgeben lassen
```

```python
# mit solve auf n Auflösen lassen
# nicht vergessen MINUS 676 einzugeben!
solve(n**2 + n + (2*(n**2/2 + n/2))/n -676,n)
```

```python
# Anstatt das Ergebnis abzutippen kann man auch den Sum Ausdruck in die solve Funktion geben
# doit() nicht vergessen, sosnt bekommt man kein Ergebnis.
# Wenn ihr Zeit habt macht beide Varianten um sicher zu gehen
solve(Sum(((2*(n*k+k))/n),(k,1,n)).doit() - 676 )
```

#### Ergebnis: n=25

## Aufgabe 5
#### Welche der folgenden Aussagen sind wahr?
#### A. $3n ∈ O(2n)$
#### B. $2n ∈ O(3n)$
#### C. $ 2^{n+42} ∈ O(3^n)$
#### D. $ (n^4 + n^3 + n^2 + n)^4 ∈ O(n^{10})$
#### E. $ \prod_{k=0}^{3} (n + k) ∈ O(n^3) $
#### F. $ \prod_{k=1}^{3} (n^k + k) ∈ O(n^3) $

```python
#Schranke: |a| ≤ C für alle a ∈ A
#bedeutet die Schranke muss größer/gleich als der Betrag jedes Elementes von A sein.
```

```python
# Die wichtigsten Regeln: S. 488 unten
# 1. Ein konstanter Faktor ändert nicht die Ordnung eines Ausdrucks
# Bsp: 4*n**3 ∈ O(n**3)     oder   2*n**2 ∈ O(n**2)

# 2. Die Ordnung der Summe zweier Ausdrücke ist die größere der Ordnungen der Summanden. 
# Bsp: n**3 + n**2 ∈ O(n**3)

# verschieden Audrücke, geordnet nach deren Wachstumsverhalten von langsam bis schnell:
# - Konstante (langsam bzw kein Wachstumsverhalten)
# - log(n)
# - n
# - n*ln(n) -> quasilinear
# - n^x
# - n!
# - x^n (schnell)

#siehe Tabelle S.491
```

```python
# A: wahr
# Ein konstanter Faktor ändert nicht die Ordnung eines Ausdrucks

# B: wahr
# Ein konstanter Faktor ändert nicht die Ordnung eines Ausdrucks

# C: wahr
# die +42 im ersten Exponenten ist vernachlässigbar da n sowieso gegen unendlich geht
# hier zählt nur die Basis

# D: falsch 
# da (n**4)**4 = n**16 viel mehr steigt als n**10
```

```python
# E: ist falsch 
# per Hand ausmultiplizieren: (n+0) * (n+1) * (n+2) * (n+3) : n**4 + 6*n**3 + 11*n**2 + 6*n
# Die Ordnung der Summe zweier Ausdrücke ist die größere der Ordnungen der Summanden. 
# n**4 steigt schneller als n**3
```

```python
# F: ist falsch 
# per Hand ausmultiplizieren: (n**1 +1) * (n**2 +2) * (n**3 +3)  : n**6 + n**5 + 2*n**4 + 5*n**3 + 3*n**2 + 6*n + 6 
# Die Ordnung der Summe zweier Ausdrücke ist die größere der Ordnungen der Summanden. 
# n**6 steigt schneller als n**3
```

```python
# Die Produkte müssen natürlich nicht immer komplet ausmultipliziert werden
# seid aber sicher den größten Exponenten gefunden zu haben!
```

## Aufgabe 5 mit Rechner
#### Welche der folgenden Aussagen sind wahr?
#### A. $3n ∈ O(2n)$
#### B. $2n ∈ O(3n)$
#### C. $ 2^{n+42} ∈ O(3^n)$
#### D. $ (n^4 + n^3 + n^2 + n)^4 ∈ O(n^{10})$
#### E. $ \prod_{k=0}^{3} (n + k) ∈ O(n^3) $
#### F. $ \prod_{k=1}^{3} (n^k + k) ∈ O(n^3) $

```python
n = symbols("n")
print("A: " , limit(3*n / (2*n) , n , oo))
# A: wahr
print("B: " , limit(2*n / (3*n) , n , oo))
# B: wahr
print("C: " , limit(2**(n+42) / 3**n,n,oo))
# C: wahr
print("D: " , limit((n**4 + n**3 + n**2 + n)**4 / n**10 , n , oo))
# D: ist falsch, da (n**4)**4 = n**16 viel mehr steigt als n**10
```

```python
# Zu E:
k = symbols("k")
z1 = Product( n + k , (k,0,3)).doit()
z1.expand()
```

```python
print("E: " , limit( z1 / n**3 , n , oo))
# E: ist falsch (Vereinfacht hätte auch n**4 / n**3 gereicht)
```

```python
# Zu F:
k = symbols("k")
z2 = Product( n**k + k , (k,1,3)).doit()
z2.expand()
```

```python
print("F: " , limit( z2 / n**3 , n , oo))
# F: ist falsch (Vereinfacht hätte auch n**6 / n**3 gereicht)
```

```python
# Bzw: Mit der in-O-Notation
# A:
limit(3*n in O(2*n) , n , oo)
```

```python
# B: 
limit(2*n in O(3*n) , n , oo)
```

```python
# C: 
limit(2**(n+42) in O(3**n) , n , oo)
```

```python
limit((n**4 + n**3 + n**2 + n)**4 in O(n**10) , n , oo)
```
