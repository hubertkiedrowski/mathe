```python
def primen(p):
    c = 0
    A = set()
    for x in range (p**2):
        if isprime(x):
            c += 1
            A.add((x))
            if c == p:
                break
    return A
```

```python
def prod(a,b):
    c = set()
    for x in a:
        for y in b:
            if x!=y:
                c.add((x*y))
    return c
```
